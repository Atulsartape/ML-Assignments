```python
dummy_cols = pd.get_dummies(df6['diagnosis'],drop_first=True,dtype=int)
```

```python
final0 = pd.concat([df6,dummy_cols], axis= 'columns')
```

```python
final0.head()
```

```python
final = final0.drop('diagnosis',axis='columns')
```

```python
final.head()
```

```python
X = final.drop('M',axis='columns')
y = final['M']
```

```python
from sklearn.preprocessing import StandardScaler 
scalling = StandardScaler()
```

```python
X_scaled = scalling.fit_transform(X)
```

```python
X_scaled
```

```python
from sklearn.model_selection import train_test_split 
X_scaled_train , X_scaled_test,y_train , y_test = train_test_split(X_scaled,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_scaled_train,y_train)
```

```python
model.score(X_scaled_test,y_test)
```

```python
y_pred = model.predict(X_scaled_test)
```

```python
from sklearn.metrics import confusion_matrix 
```

```python
cn = confusion_matrix(y_pred,y_test)
cn
```

```python
import seaborn as sns 
sns.heatmap(cn,annot=True)
```
